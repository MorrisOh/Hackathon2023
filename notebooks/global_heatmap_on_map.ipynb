```python
import pandas as pd
import plotly.express as px
import numpy as np
```

```python
# Data with latitude/longitude and values
df = pd.read_csv('https://raw.githubusercontent.com/R-CoderDotCom/data/main/sample_datasets/population_galicia.csv')

fig = px.density_mapbox(df, lat = 'latitude', lon = 'longitude', z = 'tot_pob',
                        radius = 7,
                        center = dict(lat = 42.83, lon = -8.35),
                        zoom = 6,
                        mapbox_style = 'open-street-map',
                        color_continuous_scale = 'rainbow',
                        opacity = 0.5)
fig.show()
```

```python
def add_people(data, location, count, scale=0.0005):
    center = location #(49.122677134296985, 9.210893272145562)
    for _ in range(count):
        latitude = np.random.normal(loc=center[0], scale=scale)
        longitude = np.random.normal(loc=center[1], scale=scale)
        total_count = np.random.randint(low=0, high=1_000)
        data["latitude"].append(latitude)
        data["longitude"].append(longitude)
        data["total_count"].append(total_count)
```

```python
data = {"latitude":[],"longitude":[],"total_count":[]}
add_people(data, (49.122677134296985, 9.210893272145562), count=500, scale=0.0003) # HHN
add_people(data, (49.142289135105344, 9.218618576154578), count=100, scale=0.0001) # Marktplatz
add_people(data, (49.141373647233955, 9.219714315371085), count=300, scale=0.00025) # Kiliansplatz
add_people(data, (49.142992650904205, 9.220788558862484), count=100, scale=0.0002) # Hafenmarkt
add_people(data, (49.142218433619234, 9.216242296607117), count=100, scale=0.0001) # Theaterschiff
add_people(data, (49.142662000174234, 9.221376024101854), count=15, scale=0.00001) # Passage
add_people(data, (49.14169421957764, 9.220035764738109), count=15, scale=0.00005) # Straßenbahn
add_people(data, (49.14277293301609, 9.208288712000648), count=100, scale=0.0005) # Bahnhof

df = pd.DataFrame(data=data)
```

```python
fig = px.density_mapbox(df, lat = 'latitude', lon = 'longitude', z = 'total_count',
                        radius = 7,
                        center = dict(lat = 49.142289135105344, lon=9.218618576154578),
                        zoom = 15,
                        mapbox_style = 'open-street-map',
                        color_continuous_scale = 'rainbow',
                        opacity = 0.5)
fig.show()
```
